# Checking the accuracy of the model for different value of b

```python
#b is the threshold value that can take up a value as the no. of features present.
#Here, b can take up a total of 32 values (0-31)
#The accuracy will be calculated for each value of b, unlike what we have done above, where we were assuming a random value of b.
```

```python
for b in range(32):
    total=0
    count=0
    for i in range(len(X_binarised_Train)):
        if(np.sum(X_binarised_Train[i,:])>=b):
            predicted=1
        else:
            predicted=0

        if(Y_train.iloc[i]==1):
            actual=1
        else:
            actual=0
        if(predicted==actual):
            count=count+1
        total=total+1
    accuracy=(count/total)*100
    print("For Threshold Value: ",b," accuracy determined is: ",accuracy)
```
